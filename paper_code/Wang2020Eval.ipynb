```python
import torch
import numpy as np
import os
from PIL import Image
from torchvision import transforms
from src.wrappers import (wang2020_setup, wang2020_pred) # Assuming Wang2020 is in 'src/wrappers.py'

# Load the Wang2020 model
# wang2020_model = wang2020_setup("blur_jpg_prob0.1.pth", cuda=False) # using gpu

# Load the finetuned ADM model
# finetuned_adm_model = wang2020_setup("/content/model_epoch_best.pth", cuda=False)

# Load the finetuned DDPM model
finetuned_ddpm_model = wang2020_setup("/content/model_epoch_best.pth", cuda=False)

# Preprocessing
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
])
```

```python
def predict_images(model, img_path, num_images):
    """Runs predictions on all images in a directory."""
    files = sorted(os.listdir(img_path))[:num_images]
    preds = np.zeros((num_images,))
    file_names = np.zeros((num_images,))

    for i, file in enumerate(files):
        image = Image.open(os.path.join(img_path, file)).convert("RGB")
        image_tensor = transform(image).unsqueeze(0)  # Add batch dimension

        with torch.no_grad():
            pred = wang2020_pred(model, image_tensor)

        preds[i] = pred[0].item()
        file_num = files[i].split('.')[0][12:]
        file_names[i] = int(file_num)

    return preds, file_names

num_images = 1500
iddd_test_real_path = "data/individual/real/test/"
iddd_test_fake_path = "data/individual/df/test/"
iddd_train_real_path = "data/individual/real/train/"
iddd_train_fake_path = "data/individual/df/train/"

dff_test_real_path = "data/deepfakeface/real/test/"
dff_test_fake_path = "data/deepfakeface/df/test/"
dff_train_real_path = "data/deepfakeface/real/train/"
dff_train_fake_path = "data/deepfakeface/df/train/"


# Run predictions for real and fake datasets
real_preds, real_files = predict_images(finetuned_ddpm_model[0], dff_train_real_path, num_images)
fake_preds, fake_files = predict_images(finetuned_ddpm_model[0], dff_train_fake_path, num_images)

# Save results
np.savez("wang2020finetunedDDPM_dff_train_real.npz", preds=real_preds, file_names=real_files)
np.savez("wang2020finetunedDDPM_dff_train_df.npz", preds=fake_preds, file_names=fake_files)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, average_precision_score

# Load predictions
real_data = np.load("wang2020finetunedDDPM_dff_train_real.npz")
fake_data = np.load("wang2020finetunedDDPM_dff_train_df.npz")

# Generate labels (0 for real, 1 for fake)
y_true = np.concatenate([np.zeros(num_images), np.ones(num_images)])

# Apply threshold (<0.5 is real, >=0.5 is fake)
y_pred = np.concatenate([real_data["preds"], fake_data["preds"]])
y_pred = (y_pred > 0.5).astype(int)

# Compute evaluation metrics
print("Accuracy:", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred))
print("Recall:", recall_score(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))
print("Average Precision:", average_precision_score(y_true, y_pred))

# Test only fake images
y_true = np.ones(num_images)
y_pred = (fake_data["preds"] > 0.5).astype(int)
print("Accuracy (only fake):", accuracy_score(y_true, y_pred))

# Test only real images
y_true = np.ones(num_images)
y_pred = (real_data["preds"] <= 0.5).astype(int)
print("Accuracy (only real):", accuracy_score(y_true, y_pred))
```
